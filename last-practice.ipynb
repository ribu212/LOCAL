```python
# Webスクレイピングに最低限必要なライブラリ
import requests
from bs4 import BeautifulSoup
import time

# アクセスしたいWebサイトのURL
url = "https://github.com/orgs/google/repositories"
```

```python
page_number = 1
titles = []

while True:
    # Webサーバにリクエストを出す．レスポンスを変数に格納しておく
    response = requests.get(f"{url}?page={page_number}")

    if response.status_code == 200:
        # HTMLソースをBeautifulSoupオブジェクトに変換する
        soup = BeautifulSoup(response.text, 'html.parser')
        time.sleep(1)
        # すべてのリポジトリ名を取り出す
        title_list = soup.find_all('a', class_='Text__StyledText-sc-17v1xeu-0 hWqAbU TitleHeader-module__inline--rL27T Title-module__anchor--SyQM6')
        # テキストのみにする.値がない場合はNoneを格納
        titles.extend([a.get_text(strip=True) if a else None for a in title_list])

        # 次のページのリンクを取得
        next_page_link = soup.find('a', class_='Text__StyledText-sc-17v1xeu-0 hWqAbU TitleHeader-module__inline--rL27T Title-module__anchor--SyQM6')

        if next_page_link:
            # 次のページがある場合はそのページに移動
            page_number += 1

            time.sleep(1)
        else:
            # 次のページがない場合は終了
            break
    else:
        break

print(titles)
```

```python
page_number = 1
languages = []

while True:
    # Webサーバにリクエストを出す．レスポンスを変数に格納しておく
    response = requests.get(f"{url}?page={page_number}")

    if response.status_code == 200:
        # HTMLソースをBeautifulSoupオブジェクトに変換する
        soup = BeautifulSoup(response.text, 'html.parser')
        time.sleep(1)
        # すべての言語を取り出す
        language_list = soup.find_all('span', class_='Text__StyledText-sc-17v1xeu-0 dNtsYW')
        # テキストのみにする。値がない場合はNoneを格納
        languages.extend([b.get_text(strip=True) if b else None for b in language_list])

        # 次のページのリンクを取得
        next_page_link = soup.find('span', class_='Text__StyledText-sc-17v1xeu-0 dNtsYW')

        if next_page_link:
            # 次のページがある場合はそのページに移動
            page_number += 1
        else:
            # 次のページがない場合は終了
            break
    else:
        break

print(languages)
```

```python
import re
page_number = 1
stars = []

while True:
    # Webサーバにリクエストを出す．レスポンスを変数に格納しておく
    response = requests.get(f"{url}?page={page_number}")

    if response.status_code == 200:
        # HTMLソースをBeautifulSoupオブジェクトに変換する
        soup = BeautifulSoup(response.text, 'html.parser')
        time.sleep(1)
        # すべてのスターを取り出す
        pattern = re.compile(r'/google/.+/stargazers$')
        star_list = soup.find_all('a', href=pattern)

        # 値がない場合にはNoneを格納
        stars.extend([c.get_text(strip=True) if c else None for c in star_list])

        # 次のページのリンクを取得
        next_page_link = soup.find('a', href=pattern)

        if next_page_link:
            #次のページがある場合はそのページに移動
            page_number += 1
        else:
            # 次のページがない場合は終了
            break
    else:
        break

print(stars)
```

```python
print(len(titles))
print(len(languages))
print(len(stars))
```

```python
#3つのリストを1つのリストに格納
repositories_list = list(zip(titles, languages, stars))
print(repositories_list)
```

```python
# DBに接続する
con = sqlite3.connect(path + db_name)

# SQLを実行するためのオブジェクトを取得
cur = con.cursor()

# データを挿入するSQL
# INSERT INTO テーブル名 VALUES (値1, 値2, ...);
sql_insert_many = "INSERT INTO repositories VALUES (?, ?, ?);"

# 4．SQLを実行
cur.executemany(sql_insert_many, repositories_list)

# 5．コミット処理（データ操作を反映させる）
con.commit()

# 6．DBへの接続を閉じる
con.close()
```
